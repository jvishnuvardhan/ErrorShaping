```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Plot batch losses 

```
def get_list_from_file(file_name):

  with open(file_name, 'r') as f:
    batch_losses = f.readlines()
  f.close()
  batch_losses = [float(element[:8]) for element in batch_losses]
  return batch_losses

def plot_losses_batch(losses_ce_batch, losses_cees_batch, losses_ceessv_batch, xaxis_n):
  
  x = np.arange(xaxis_n)
  fig = plt.figure(dpi=100)
  line1 = plt.plot(x, losses_ce_batch[:xaxis_n], 'r', label='ce') 
  line2= plt.plot(x, losses_cees_batch[:xaxis_n], 'b', label='cees') 
  line3= plt.plot(x, losses_ceessv_batch[:xaxis_n], 'g', label='ceessv') 
  plt.xlabel("batch")
  plt.ylabel("loss")
  plt.legend()
  plt.show()

losses_ce_batch = get_list_from_file("losses_ce_batch.txt")
print("losses_ce_batch mean:    ", np.mean(losses_ce_batch))

losses_cees_batch = get_list_from_file("losses_cees_batch.txt")
print("losses_cees_batch mean:  ", np.mean(losses_cees_batch))

losses_ceessv_batch = get_list_from_file("losses_ceessv_batch.txt")
print("losses_ceessv_batch mean:", np.mean(losses_ceessv_batch))

plot_losses_batch(losses_ce_batch, losses_cees_batch, losses_ceessv_batch, len(losses_ce_batch))
plot_losses_batch(losses_ce_batch, losses_cees_batch, losses_ceessv_batch, 150)
```

## Plot 3D batch losses distribution

```
losses_ce_batch = get_list_from_file("losses_ce_batch.txt")
print("losses_ce_batch mean:    ", np.mean(losses_ce_batch))
print("losses_ce_batch mean of the first 5000:    ", np.mean(losses_ce_batch[150]))

losses_cees_batch = get_list_from_file("losses_cees_batch.txt")
print("losses_cees_batch mean:  ", np.mean(losses_cees_batch))
print("losses_cees_batch mean of the first 5000:    ", np.mean(losses_cees_batch[150]))

losses_ceessv_batch = get_list_from_file("losses_ceessv_batch.txt")
print("losses_ceessv_batch mean:", np.mean(losses_ceessv_batch))
print("losses_ceessv_batch mean of the first 5000:    ", np.mean(losses_ceessv_batch[150]))
```

```
def get_array_from_file(file_name, sample_length=1000):
  with open(file_name, 'r') as f:
    content = f.readlines()
    points_n = len(content)
  f.close()
  points = [float(element[:8]) for element in content]
  #print(type(points))
  #points = points[:2000]

  arr = []
  sample_n = points_n - sample_length
  for i in range(sample_n):
    sample = points[i:i+sample_length]
    if len(sample)==sample_length:
      arr.append(sample)
  return np.array([np.array(xi) for xi in arr]), sample_n

def ErrorArray_to_HistArray(error_array, bin_n=100):
  pd_arr = []
  edges = []
  for i in range(error_array.shape[0]):
    counts, edge=np.histogram(error_array[i], range=(0, 0.4), bins=bin_n) #pd for probability density
    pd = counts/sum(counts)
    pd_arr.append(pd)
    edges.append(edge)
  
  edge = edges[0]
  edge_middlized = [(edge[i]+edge[i+1])/2 for i in range(len(edge)-1)]
  return np.array(pd_arr), np.array(edge_middlized)

def plot_errors(pd_array, end, edge, title):
  fig = plt.figure(dpi=150)
  ax = fig.add_subplot(111, projection='3d')
  x_data = np.tile(edge, pd_array.shape[0]) # x_data.shape (200,)
  y_data = np.repeat(np.arange(pd_array.shape[1]), end).flatten() # y_data.shape (200,)
  z_data = pd_array.flatten() # z_data.shape (5900000,)
  ax.bar3d(x_data, y_data, np.zeros(len(z_data)),dx=0.01, dy=1, dz=z_data)
  ax.set_zlim(0, 1)
  ax.set_xlabel('errors') 
  ax.set_ylabel('sample_n = total sample - sample') 
  plt.title(title)
  ax.set_zlabel('probability density')
  plt.show()
```

```


sample_n = 5000
sample_length= 60000 - sample_n
error_array, end = get_array_from_file("losses_ce_sample.txt", sample_length=sample_length)
pd_array, edge = ErrorArray_to_HistArray(error_array)  
plot_errors(pd_array, end, edge, "losses_ce_sample")
```

```
sample_n = 5000
sample_length= 60000 - sample_n
error_array, end = get_array_from_file("losses_cees_sample.txt", sample_length=sample_length)
pd_array, edge = ErrorArray_to_HistArray(error_array)  
plot_errors(pd_array, end, edge, "losses_cees_sample")
```

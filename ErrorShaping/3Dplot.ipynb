```
import numpy as np
import matplotlib.pyplot as plt
```

```
def get_epoch_from_batch(points):
  batch_n_each_epoc = 4
  loop_n = len(points)//batch_n_each_epoc
  losses_epoch = []
  for i in range(loop_n):
    losses_epoch.append(np.mean(points[batch_n_each_epoc*i:batch_n_each_epoc*(i+1)]))
  return losses_epoch


ce = get_batch_from_file("ce_losses.txt")
ce = get_epoch_from_batch(ce)
cees = get_batch_from_file("cees_losses.txt")
cees = get_epoch_from_batch(cees)
ceessv = get_batch_from_file("ceessv_losses.txt")
ceessv = get_epoch_from_batch(ceessv)
print(len(ce))
print(len(cees))
print(len(ceessv))

l = len(ce)
x = np.arange(l)
fig = plt.figure(dpi=150)
line1 = plt.plot(x, ce[:l], 'r', label='ce') 
line2= plt.plot(x, cees[:l], 'b', label='cees') 
line3= plt.plot(x, ceessv[:l], 'g', label='ceessv') 
plt.legend()
plt.show()
```

```
def get_arr_from_list(points, sample_n=100, sample_length=500):

  gap = len(points)//sample_n
  arr = []
  for i in range(sample_n):
    sample = points[gap*i: gap*i+sample_length]
    if len(sample)==sample_length:
      arr.append(sample)

  return np.array([np.array(xi) for xi in arr])

sample_n = 3
sample_length= 5
arr  = get_arr_from_list(ce, sample_n=sample_n, sample_length=sample_length)
arr
```

```
import numpy as np

def get_batch_from_file(file_name):

  with open(file_name, 'r') as f:
    content = f.readlines()
  f.close()
  content = [float(element[:8]) for element in content]
  print("len(content):", len(content))
  batch_n = 32
  loop_n = len(content)//batch_n
  losses_batch = []
  for i in range(loop_n):
    losses_batch.append(np.mean(content[batch_n*i:batch_n*(i+1)]))
  print("len(losses_batch):", len(losses_batch))
  return losses_batch

ce = get_batch_from_file("ce_losses.txt")
cees = get_batch_from_file("cees_losses.txt")
ceessv = get_batch_from_file("ceessv_losses.txt")
print(len(ce))
print(len(cees))
print(len(ceessv))

l = len(ce)
x = np.arange(l)
fig = plt.figure(dpi=150)
line1 = plt.plot(x, ce[:l], 'r', label='ce') 
line2= plt.plot(x, cees[:l], 'b', label='cees') 
line3= plt.plot(x, ceessv[:l], 'g', label='ceessv') 
plt.legend()
plt.show()
```
